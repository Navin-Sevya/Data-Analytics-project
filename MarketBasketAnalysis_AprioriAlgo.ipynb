```python
#Import Libraries----------
import numpy as np 
import pandas as pd 
from mlxtend.frequent_patterns import apriori, association_rules 
```

```python
#Loading and exploring the data-----------------
#Loading the Data 
data = pd.read_excel('D:\SKILLEDGES PUNE\Datasets/Online_Retail.xlsx') 
data
```

```python
#Cleaning the Data-----------------
#Identifying missing values:
'''Is there any missing values across columns'''
data.isnull().any()
```

```python
'''How many missing values are there across each column'''
data.isnull().sum()
```

```python
data.info()
```

```python
# Dropping all transactions which were done on credit ('C')
data['InvoiceNo'] = data['InvoiceNo'].astype('str') 
data = data[~data['InvoiceNo'].str.contains('C')]
```

```python
# Stripping extra spaces in the description 
data['Description'] = data['Description'].str.strip()
data
```

```python
#Splitting the data according to the region of transaction-------
```

```python
# Transactions done in France 
basket_France = (data[data['Country'] =="France"] 
        .groupby(['InvoiceNo', 'Description'])['Quantity'] 
        .sum().unstack().reset_index().fillna(0) 
        .set_index('InvoiceNo'))
basket_France
```

```python

# Transactions done in the United Kingdom 
basket_UK = (data[data['Country'] =="United Kingdom"] 
        .groupby(['InvoiceNo', 'Description'])['Quantity'] 
        .sum().unstack().reset_index().fillna(0) 
        .set_index('InvoiceNo'))
basket_UK
```

```python

# Transactions done in Portugal 
basket_Por = (data[data['Country'] =="Portugal"] 
        .groupby(['InvoiceNo', 'Description'])['Quantity'] 
        .sum().unstack().reset_index().fillna(0) 
        .set_index('InvoiceNo')) 
basket_Por
```

```python
basket_Sweden = (data[data['Country'] =="Sweden"] 
        .groupby(['InvoiceNo', 'Description'])['Quantity'] 
        .sum().unstack().reset_index().fillna(0) 
        .set_index('InvoiceNo'))
basket_Sweden
```

```python
#Hot encoding the Data------------
```

```python
# Defining the hot encoding function to make the data suitable 
# for the concerned libraries
```

```python
def hot_encode(x): 
    if(x<= 0): 
        return 0
    if(x>= 1): 
        return 1
```

```python
# Encoding the datasets
```

```python
basket_encoded_1 = basket_France.applymap(hot_encode) 
basket_France = basket_encoded_1
basket_France
```

```python
basket_encoded_2 = basket_UK.applymap(hot_encode) 
basket_UK = basket_encoded_2
basket_UK
```

```python
basket_encoded_3 = basket_Por.applymap(hot_encode) 
basket_Por = basket_encoded_3
basket_Por
```

```python
basket_encoded_4 = basket_Sweden.applymap(hot_encode) 
basket_Sweden = basket_encoded_4
basket_Sweden
```

```python
#Building the models and analyzing the results-----------------
```

```python
#France:
# Building the model 
frq_items = apriori(basket_France, min_support = 0.1, use_colnames = True) 
frq_items
```

```python
# Collecting the inferred rules in a dataframe 
rules = association_rules(frq_items, metric ="lift", min_threshold = 1) 
print(rules.head())
```

```python
France_rules=pd.DataFrame(rules)
France_rules
```

```python
#Portugal
frq_items = apriori(basket_Por, min_support = 0.15, use_colnames = True) 
rules = association_rules(frq_items, metric ="lift", min_threshold = 1) 
print(rules.head()) 
```

```python
Portugal_rules=pd.DataFrame(rules)
Portugal_rules
```

```python
#Sweden
frq_items = apriori(basket_Sweden, min_support = 0.10, use_colnames = True) 
rules = association_rules(frq_items, metric ="lift", min_threshold = 1) 
print(rules.head()) 
```

```python
Sweden_rules=pd.DataFrame(rules)
Sweden_rules
```

```python
#UK
frq_items = apriori(basket_UK, min_support = 0.05, use_colnames = True) 
rules = association_rules(frq_items, metric ="lift", min_threshold = 1) 
print(rules.head()) 
```

```python
UK_rules=pd.DataFrame(rules)
UK_rules
```
